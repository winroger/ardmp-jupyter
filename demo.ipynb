```python
from collections import Counter
from pathlib import Path
import re

import matplotlib.pyplot as plt

data_path = Path("3f0bc6a4-92de-44c6-b4ec-6bcb1a61528f/data/dataset.ttl")
ttl_text = data_path.read_text(encoding="utf-8")

building_blocks = re.findall(
    r"(Bui:[^\n]+.*?gnd:dateOfProduction\s+\"([^\"]+)\".*?)\.",
    ttl_text,
    flags=re.DOTALL,
 )

years = []
for _, year in building_blocks:
    year = year.strip()
    if year:
        years.append(year)

year_counts = Counter(years)
sorted_years = sorted(year_counts, key=lambda value: int(value))
counts = [year_counts[year] for year in sorted_years]

plt.figure(figsize=(10, 5))
plt.bar(sorted_years, counts, color="#2f6b8a")
plt.title("Buildings by gnd:dateOfProduction")
plt.xlabel("Date of production")
plt.ylabel("Number of buildings")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

year_counts
```
